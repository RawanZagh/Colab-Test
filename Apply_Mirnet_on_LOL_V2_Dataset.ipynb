Resize images in the Normal Dataset

```python
from PIL import Image
import os

# Define the input and output directories
input_directory = "/content/drive/My Drive/Normal"
output_directory = "/content/drive/My Drive/Normal-S"

# Desired size
desired_size = (256, 256)

# Ensure the output directory exists
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterate through files in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith((".jpg", ".png")):
        # Open the image
        with Image.open(os.path.join(input_directory, filename)) as img:
            # Resize the image
            img = img.resize(desired_size, Image.ANTIALIAS)

            # Save the resized image to the output directory
            img.save(os.path.join(output_directory, filename))

print("Resizing complete.")
```

```python
import numpy as np # array manipulation
from huggingface_hub import from_pretrained_keras # download the model
import keras # deep learning
from PIL import Image # Image processing
```

```python
model = from_pretrained_keras("keras-io/lowlight-enhance-mirnet", compile=False)
```

```python
from keras.preprocessing import image
import os
import numpy as np
from PIL import Image

# Define the input directory and output directory
input_directory = "/content/drive/My Drive/Low-S"
output_directory = "/content/drive/My Drive/Result"

# Load the enhancement model (replace 'model' with your actual model)
# model = ...

# Iterate through files in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith((".jpg", ".png")):
        # Open and load the image
        img = image.load_img(os.path.join(input_directory, filename), target_size=(256, 256))

        # Convert the image to a NumPy array
        img_array = image.img_to_array(img)
        img_array = img_array.astype('float32') / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        # Perform image enhancement using the model
        output = model.predict(img_array)

        # Clip and reshape the enhanced image
        output_image = output[0] * 255.0
        output_image = output_image.clip(0, 255)
        output_image = output_image.reshape((np.shape(output_image)[0], np.shape(output_image)[1], 3))

        # Convert the NumPy array to a PIL image
        output_image = Image.fromarray(output_image.astype('uint8'), 'RGB')

        # Save the enhanced image to the output directory
        output_image.save(os.path.join(output_directory, filename))

print("Image processing complete.")
```
